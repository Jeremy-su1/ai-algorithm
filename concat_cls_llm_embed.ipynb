```python
# -*- coding: utf-8 -*-
import os
import torch
from transformers import AutoConfig, AutoModelForCausalLM, AutoTokenizer, DataCollatorWithPadding
from tqdm import trange
from datasets import Dataset, DatasetDict
import pandas as pd
import numpy as np
import re
from torch.cuda.amp import autocast

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # 다중 공백을 단일 공백으로 변환
    text = re.sub(r'<[^>]+>', '', text)  # HTML 태그 제거
    text = text.strip()
    return text

def get_sequence_lengths(tokenizer, sents):
    lengths = []
    for sent in sents:
        tokenized = tokenizer.encode(sent, add_special_tokens=False)
        lengths.append(len(tokenized))
    return lengths

def rep_extract(task, mode, device, sents, multiclass_labels, multilabels, tokenizer, model, max_len, step):
    sents_reps = []
    for idx in trange(0, len(sents), step):
        idx_end = idx + step
        if idx_end > len(sents):
            idx_end = len(sents)
        sents_batch = sents[idx: idx_end]
        multiclass_batch = multiclass_labels[idx: idx_end]
        multilabels_batch = multilabels[idx: idx_end]

        sents_batch_encoding = tokenizer(sents_batch, return_tensors='pt', max_length=max_len, padding="max_length", truncation=True)
        sents_batch_encoding = sents_batch_encoding.to(device)

        with autocast():
            with torch.no_grad():
                batch_outputs = model(**sents_batch_encoding)
                reps_batch_5L = []
                for layer in range(-1, -6, -1):
                    reps_batch_5L.append(torch.mean(batch_outputs.hidden_states[layer], axis=1))
                reps_batch_5L = torch.stack(reps_batch_5L, axis=1)

        sents_reps.append(reps_batch_5L.cpu())
    sents_reps = torch.cat(sents_reps)

    # Multiclass와 Multilabel 데이터를 Tensor로 변환, None 값을 -1로 대체
    multiclass_labels_tensor = torch.tensor([label if label is not None else -1 for label in multiclass_labels])
    multilabels_tensor = torch.tensor(multilabels, dtype=torch.float32)

    print(f"Sentence embeddings shape: {sents_reps.shape}")
    print(f"Multiclass labels shape: {multiclass_labels_tensor.shape}")
    print(f"Multilabels shape: {multilabels_tensor.shape}")

    # 데이터 저장
    path = f'/content/drive/MyDrive/LLMEmbed/UnifiedModel/{task}/dataset_tensor/'
    if not os.path.exists(path):
        os.makedirs(path)
    torch.save(sents_reps.to('cpu'), path + f'{mode}_sents.pt')
    torch.save(multiclass_labels_tensor, path + f'{mode}_multiclass_labels.pt')
    torch.save(multilabels_tensor, path + f'{mode}_multilabels.pt')

# 실행 코드
if __name__ == '__main__':
    cuda_no = 0
    device = f'cuda:{cuda_no}' if torch.cuda.is_available() else 'cpu'
    model_id = "meta-llama/Llama-3.2-1B"
    access_token = "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"
    tokenizer = AutoTokenizer.from_pretrained(model_id, use_auth_token="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")
    # tokenizer.pad_token = "[PAD]"
    # tokenizer.padding_side = "right"
    tokenizer.pad_token = tokenizer.eos_token

    config_kwargs = {
        "trust_remote_code": True,
        "cache_dir": None,
        "revision": 'main',
        "use_auth_token": None,
        "output_hidden_states": True
    }
    model_config = AutoConfig.from_pretrained(model_id, **config_kwargs, token=access_token)
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        trust_remote_code=True,
        config=model_config,
        device_map=device,
        torch_dtype=torch.float16,
        use_auth_token=access_token)
    model.eval()

    # 레이블 컬럼 목록
    label_cols = ['Algorithms', 'Backend', 'Data Science', 'Databases', 'Dev Tools', 'Frontend', 'Mobile', 'Systems', 'iOS/macOS']

    # multi-class 데이터셋 추출
    train_df = pd.DataFrame(sec_dataset['train'])
    test_df = pd.DataFrame(sec_dataset['test'])
    train_sents = train_df['title_content'].apply(clean_text).tolist()
    test_sents = test_df['title_content'].apply(clean_text).tolist()
    train_labels = train_df['label'].tolist()
    test_labels = test_df['label'].tolist()

    # multi-label 데이터셋 추출
    multi_label_train_df = pd.DataFrame(sof_dataset['train'])
    multi_label_test_df = pd.DataFrame(sof_dataset['test'])
    multi_label_train_sents = multi_label_train_df['Body'].apply(clean_text).tolist()
    multi_label_test_sents = multi_label_test_df['Body'].apply(clean_text).tolist()
    multi_label_train_labels = multi_label_train_df[label_cols].values.tolist()
    multi_label_test_labels = multi_label_test_df[label_cols].values.tolist()

    # 통합된 title과 레이블 생성
    combined_train_sents = train_sents + multi_label_train_sents
    combined_test_sents = test_sents + multi_label_test_sents
    combined_train_multiclass_labels = train_labels + [None] * len(multi_label_train_sents)
    combined_test_multiclass_labels = test_labels + [None] * len(multi_label_test_sents)
    combined_train_multilabels = [[0] * len(label_cols)] * len(train_sents) + multi_label_train_labels
    combined_test_multilabels = [[0] * len(label_cols)] * len(test_sents) + multi_label_test_labels

    # 시퀀스 길이 분석 및 적절한 max_len 결정
    all_lengths = get_sequence_lengths(tokenizer, combined_train_sents + combined_test_sents)
    max_len = int(np.percentile(all_lengths, 95)) + 10  # 95번째 퍼센타일에 여유분 추가
    print(f"Dynamic max_len set to: {max_len}")

    # 통합 데이터셋 처리
    rep_extract('llama3_2', 'train', device, combined_train_sents, combined_train_multiclass_labels, combined_train_multilabels, tokenizer, model, max_len, step=5)
    rep_extract('llama3_2', 'test', device, combined_test_sents, combined_test_multiclass_labels, combined_test_multilabels, tokenizer, model, max_len, step=5)
```

# **2. BERT Embedding**

```python
# -*- coding: utf-8 -*-
import os
import torch
from transformers import BertTokenizer, BertModel, DataCollatorWithPadding
from tqdm import trange
import pandas as pd
import numpy as np
import re
from torch.cuda.amp import autocast  # for mixed precision

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # 다중 공백을 단일 공백으로 변환
    text = re.sub(r'<[^>]+>', '', text)  # HTML 태그 제거
    text = text.strip()
    return text

def get_sequence_lengths(tokenizer, sents):
    lengths = []
    for sent in sents:
        tokenized = tokenizer.encode(sent, add_special_tokens=False)
        lengths.append(len(tokenized))
    return lengths

def rep_extract(task, mode, device, sents, multiclass_labels, multilabels, tokenizer, model, max_len, step):
    sents_reps = []
    for idx in trange(0, len(sents), step):
        idx_end = idx + step
        if idx_end > len(sents):
            idx_end = len(sents)
        sents_batch = sents[idx: idx_end]
        multiclass_batch = multiclass_labels[idx: idx_end]
        multilabels_batch = multilabels[idx: idx_end]

        # 동적 패딩을 위한 DataCollatorWithPadding 사용
        sents_batch_encoding = tokenizer(sents_batch, return_tensors='pt', max_length=max_len, padding="max_length", truncation=True)
        sents_batch_encoding = sents_batch_encoding.to(device)

        # 혼합 정밀도 적용
        with autocast():
            with torch.no_grad():
                batch_outputs = model(**sents_batch_encoding)
                reps_batch = batch_outputs.pooler_output

        sents_reps.append(reps_batch.cpu())
    sents_reps = torch.cat(sents_reps)

    # Multiclass와 Multilabel 데이터를 Tensor로 변환, None 값을 -1로 대체
    multiclass_labels_tensor = torch.tensor([label if label is not None else -1 for label in multiclass_labels], dtype=torch.long)
    multilabels_tensor = torch.tensor(multilabels, dtype=torch.float32)

    print(f"Sentence embeddings shape: {sents_reps.shape}")
    print(f"Multiclass labels shape: {multiclass_labels_tensor.shape}")
    print(f"Multilabels shape: {multilabels_tensor.shape}")

    # 데이터 저장
    path = f'/content/drive/MyDrive/LLMEmbed/BERT/{task}/dataset_tensor/'
    if not os.path.exists(path):
        os.makedirs(path)
    torch.save(sents_reps.to('cpu'), path + f'{mode}_sents.pt')
    torch.save(multiclass_labels_tensor, path + f'{mode}_multiclass_labels.pt')
    torch.save(multilabels_tensor, path + f'{mode}_multilabels.pt')

# 실행 코드
if __name__ == '__main__':
    cuda_no = 0
    device = torch.device(f"cuda:{cuda_no}" if torch.cuda.is_available() else "cpu")
    model_path = 'google-bert/bert-large-uncased'

    tokenizer = BertTokenizer.from_pretrained(model_path)
    model = BertModel.from_pretrained(model_path).to(device)
    model.eval()

    # 레이블 컬럼 목록
    label_cols = ['Algorithms', 'Backend', 'Data Science', 'Databases', 'Dev Tools', 'Frontend', 'Mobile', 'Systems', 'iOS/macOS']

    # multi-class 데이터셋 추출
    train_df = pd.DataFrame(sec_dataset['train'])
    test_df = pd.DataFrame(sec_dataset['test'])
    train_sents = train_df['title_content'].apply(clean_text).tolist()
    test_sents = test_df['title_content'].apply(clean_text).tolist()
    train_labels = train_df['label'].tolist()
    test_labels = test_df['label'].tolist()

    # multi-label 데이터셋 추출
    multi_label_train_df = pd.DataFrame(sof_dataset['train'])
    multi_label_test_df = pd.DataFrame(sof_dataset['test'])
    multi_label_train_sents = multi_label_train_df['Body'].apply(clean_text).tolist()
    multi_label_test_sents = multi_label_test_df['Body'].apply(clean_text).tolist()
    multi_label_train_labels = multi_label_train_df[label_cols].values.tolist()
    multi_label_test_labels = multi_label_test_df[label_cols].values.tolist()

    # 통합된 title과 레이블 생성
    combined_train_sents = train_sents + multi_label_train_sents
    combined_test_sents = test_sents + multi_label_test_sents
    combined_train_multiclass_labels = train_labels + [None] * len(multi_label_train_sents)
    combined_test_multiclass_labels = test_labels + [None] * len(multi_label_test_sents)
    combined_train_multilabels = [[0] * len(label_cols)] * len(train_sents) + multi_label_train_labels
    combined_test_multilabels = [[0] * len(label_cols)] * len(test_sents) + multi_label_test_labels

    # 시퀀스 길이 분석 및 적절한 max_len 결정
    all_lengths = get_sequence_lengths(tokenizer, combined_train_sents + combined_test_sents)
    dynamic_max_len = int(np.percentile(all_lengths, 95)) + 10  # 95번째 퍼센타일에 여유분 추가
    max_len = min(512, dynamic_max_len)  # BERT의 최대 입력 길이인 512로 제한
    print(f"Dynamic max_len set to: {max_len}")


    # 통합 데이터셋 처리
    rep_extract('bert', 'train', device, combined_train_sents, combined_train_multiclass_labels, combined_train_multilabels, tokenizer, model, max_len, step=5)
    rep_extract('bert', 'test', device, combined_test_sents, combined_test_multiclass_labels, combined_test_multilabels, tokenizer, model, max_len, step=5)
```

# **3. Roberta Embedding**

```python
# -*- coding: utf-8 -*-
import os
import torch
from transformers import RobertaTokenizer, RobertaModel, DataCollatorWithPadding
from tqdm import trange
import pandas as pd
import numpy as np
import re
from torch.cuda.amp import autocast  # for mixed precision

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # 다중 공백을 단일 공백으로 변환
    text = re.sub(r'<[^>]+>', '', text)  # HTML 태그 제거
    text = text.strip()
    return text

def get_sequence_lengths(tokenizer, sents):
    lengths = []
    for sent in sents:
        tokenized = tokenizer.encode(sent, add_special_tokens=False)
        lengths.append(len(tokenized))
    return lengths

def rep_extract(task, mode, device, sents, multiclass_labels, multilabels, tokenizer, model, max_len, step):
    sents_reps = []
    for idx in trange(0, len(sents), step):
        idx_end = idx + step
        if idx_end > len(sents):
            idx_end = len(sents)
        sents_batch = sents[idx: idx_end]
        multiclass_batch = multiclass_labels[idx: idx_end]
        multilabels_batch = multilabels[idx: idx_end]

        # 동적 패딩을 위한 DataCollatorWithPadding 사용
        sents_batch_encoding = tokenizer(sents_batch, return_tensors='pt', padding='max_length', truncation=True, max_length=max_len)
        sents_batch_encoding = sents_batch_encoding.to(device)

        # 혼합 정밀도 적용
        with autocast():
            with torch.no_grad():
                batch_outputs = model(**sents_batch_encoding)
                # 첫 번째 토큰([CLS] 토큰에 해당)의 표현을 추출
                reps_batch = batch_outputs.last_hidden_state[:, 0, :]

        sents_reps.append(reps_batch.cpu())
    sents_reps = torch.cat(sents_reps)

    # Multiclass와 Multilabel 데이터를 Tensor로 변환, None 값을 -1로 대체
    multiclass_labels_tensor = torch.tensor([label if label is not None else -1 for label in multiclass_labels], dtype=torch.long)
    multilabels_tensor = torch.tensor(multilabels, dtype=torch.float32)

    print(f"Sentence embeddings shape: {sents_reps.shape}")
    print(f"Multiclass labels shape: {multiclass_labels_tensor.shape}")
    print(f"Multilabels shape: {multilabels_tensor.shape}")

    # 데이터 저장
    path = f'/content/drive/MyDrive/LLMEmbed/RoBERTa/{task}/dataset_tensor/'
    if not os.path.exists(path):
        os.makedirs(path)
    torch.save(sents_reps.to('cpu'), path + f'{mode}_sents.pt')
    torch.save(multiclass_labels_tensor, path + f'{mode}_multiclass_labels.pt')
    torch.save(multilabels_tensor, path + f'{mode}_multilabels.pt')

# 실행 코드
if __name__ == '__main__':
    cuda_no = 0
    device = torch.device(f"cuda:{cuda_no}" if torch.cuda.is_available() else "cpu")
    model_path = 'roberta-large'

    tokenizer = RobertaTokenizer.from_pretrained(model_path)
    model = RobertaModel.from_pretrained(model_path).to(device)
    model.eval()

    # 레이블 컬럼 목록
    label_cols = ['Algorithms', 'Backend', 'Data Science', 'Databases', 'Dev Tools', 'Frontend', 'Mobile', 'Systems', 'iOS/macOS']

    # multi-class 데이터셋 추출
    train_df = pd.DataFrame(sec_dataset['train'])
    test_df = pd.DataFrame(sec_dataset['test'])
    train_sents = train_df['title_content'].apply(clean_text).tolist()
    test_sents = test_df['title_content'].apply(clean_text).tolist()
    train_labels = train_df['label'].tolist()
    test_labels = test_df['label'].tolist()

    # multi-label 데이터셋 추출
    multi_label_train_df = pd.DataFrame(sof_dataset['train'])
    multi_label_test_df = pd.DataFrame(sof_dataset['test'])
    multi_label_train_sents = multi_label_train_df['Body'].apply(clean_text).tolist()
    multi_label_test_sents = multi_label_test_df['Body'].apply(clean_text).tolist()
    multi_label_train_labels = multi_label_train_df[label_cols].values.tolist()
    multi_label_test_labels = multi_label_test_df[label_cols].values.tolist()

    # 통합된 title과 레이블 생성
    combined_train_sents = train_sents + multi_label_train_sents
    combined_test_sents = test_sents + multi_label_test_sents
    combined_train_multiclass_labels = train_labels + [None] * len(multi_label_train_sents)
    combined_test_multiclass_labels = test_labels + [None] * len(multi_label_test_sents)
    combined_train_multilabels = [[0] * len(label_cols)] * len(train_sents) + multi_label_train_labels
    combined_test_multilabels = [[0] * len(label_cols)] * len(test_sents) + multi_label_test_labels

    # 시퀀스 길이 분석 및 적절한 max_len 결정
    all_lengths = get_sequence_lengths(tokenizer, combined_train_sents + combined_test_sents)
    dynamic_max_len = int(np.percentile(all_lengths, 95)) + 10  # 95번째 퍼센타일에 여유분 추가
    max_len = min(512, dynamic_max_len)  # RoBERTa의 최대 입력 길이인 512로 제한
    print(f"Dynamic max_len set to: {max_len}")

    # 통합 데이터셋 처리
    rep_extract('roberta', 'train', device, combined_train_sents, combined_train_multiclass_labels, combined_train_multilabels, tokenizer, model, max_len, step=5)
    rep_extract('roberta', 'test', device, combined_test_sents, combined_test_multiclass_labels, combined_test_multilabels, tokenizer, model, max_len, step=5)
```

# **4. Dataset 정의**

```python
import torch
from torch.utils.data import Dataset, DataLoader

class MyDataset(Dataset):

    def __init__(self, mode, l_path, b_path, r_path, is_multilabel=False):
        """
        :param mode: 'train' 또는 'test'와 같은 데이터 모드
        :param l_path: Llama 모델의 임베딩 경로
        :param b_path: BERT 모델의 임베딩 경로
        :param r_path: RoBERTa 모델의 임베딩 경로
        :param is_multilabel: True인 경우 multi-label 데이터를 불러옴
        """
        self.l_sents_reps = torch.load(l_path + f'{mode}_sents.pt')
        self.b_sents_reps = torch.load(b_path + f'{mode}_sents.pt')
        self.r_sents_reps = torch.load(r_path + f'{mode}_sents.pt')

        # multi-label 또는 multi-class 레이블 로드
        self.is_multilabel = is_multilabel
        if is_multilabel:
            self.multiclass_labels = torch.load(l_path + f'{mode}_multiclass_labels.pt')
            self.multilabels = torch.load(l_path + f'{mode}_multilabels.pt')
            self.sample_num = self.multiclass_labels.shape[0]
        else:
            self.labels = torch.load(l_path + f'{mode}_labels.pt')
            self.sample_num = self.labels.shape[0]

    def __getitem__(self, index):
        # multi-label과 multi-class 레이블을 모두 포함하여 반환
        if self.is_multilabel:
            return (self.l_sents_reps[index],
                    self.b_sents_reps[index],
                    self.r_sents_reps[index],
                    self.multiclass_labels[index],
                    self.multilabels[index])
        else:
            return (self.l_sents_reps[index],
                    self.b_sents_reps[index],
                    self.r_sents_reps[index],
                    self.labels[index])

    def __len__(self):
        return self.sample_num
```

# * Model Operation - 검증셋 테스트

```python
import numpy as np
import torch
import wandb
import evaluate
from tqdm import tqdm
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# wandb.init(project="huggingface") # Uncomment to use wandb

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def compute_metrics(predictions, references, is_multilabel):
    # Multi-label 예측 및 레퍼런스를 다루는 경우와 multi-class의 경우를 구분
    if is_multilabel:
        predictions = torch.sigmoid(predictions)
        predictions = (predictions > 0.5).int()
    else:
        predictions = torch.argmax(predictions, dim=1)

    # 텐서를 CPU로 이동시키고 NumPy 배열로 변환
    predictions = predictions.cpu().numpy()
    references = references.cpu().numpy()

    # multi-label인 경우 각 클래스의 평균 점수를 계산
    if is_multilabel:
        accuracy = accuracy_score(references, predictions)
        micro_precision, micro_recall, micro_f1, _ = precision_recall_fscore_support(references, predictions, average='micro')
        macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(references, predictions, average='macro')
        weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(references, predictions, average='weighted')

        # 평탄화된 정확도 계산
        flat_predictions = predictions.flatten()
        flat_references = references.flatten()
        flat_accuracy = accuracy_score(flat_references, flat_predictions)

        # 평가 메트릭 반환
        return {
            'flat_accuracy': flat_accuracy,
            'accuracy': accuracy,
            'macro_precision': macro_precision,
            'macro_recall': macro_recall,
            'macro_f1': macro_f1,
            'micro_precision': micro_precision,
            'micro_recall': micro_recall,
            'micro_f1': micro_f1,
            'weighted_precision': weighted_precision,
            'weighted_recall': weighted_recall,
            'weighted_f1': weighted_f1,
        }
    else:
        # multi-class 평가 메트릭
        accuracy = accuracy_score(references, predictions)
        precision, recall, f1, _ = precision_recall_fscore_support(references, predictions, average='weighted')
        return {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1': f1,
        }

def Train_and_Evaluate(dataloader_train, dataloader_val, device, model, loss_fn_multi_class, loss_fn_multi_label, optimizer):
    # Training
    model.train()
    total_train_loss = 0
    for batch in tqdm(dataloader_train):
        batch_l, batch_b, batch_r, batch_y_class, batch_y_label = [item.to(device) for item in batch]
        # batch_y_label을 float으로 변경
        batch_y_label = batch_y_label.float()

        # 모델 예측
        pred_class, pred_label = model(batch_l.float(), batch_b.float(), batch_r.float())

        # 크기 확인 (디버깅)
        print(f"pred_class shape: {pred_class.shape}, batch_y_class shape: {batch_y_class.shape}")
        print(f"pred_label shape: {pred_label.shape}, batch_y_label shape: {batch_y_label.shape}")

        # 각 손실 계산
        loss_class = loss_fn_multi_class(pred_class, batch_y_class.long())  # multi-class 손실
        loss_label = loss_fn_multi_label(pred_label, batch_y_label.float())  # multi-label 손실
        loss = loss_class + loss_label
        total_train_loss += loss.item()

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    avg_train_loss = total_train_loss / len(dataloader_train)

    # Validation
    model.eval()
    total_val_loss = 0
    all_predictions_class = []
    all_references_class = []
    all_predictions_label = []
    all_references_label = []

    with torch.no_grad():
        for batch in tqdm(dataloader_val):
            batch_l, batch_b, batch_r, batch_y_class, batch_y_label = [item.to(device) for item in batch]

            # 모델 예측
            pred_class, pred_label = model(batch_l.float(), batch_b.float(), batch_r.float())

            # 각 손실 계산
            loss_class = loss_fn_multi_class(pred_class, batch_y_class)  # multi-class 손실
            loss_label = loss_fn_multi_label(pred_label, batch_y_label.float())  # multi-label 손실
            loss = loss_class + loss_label
            total_val_loss += loss.item()

            # 예측값 저장
            all_predictions_class.append(pred_class)
            all_references_class.append(batch_y_class)
            all_predictions_label.append(pred_label)
            all_references_label.append(batch_y_label)

    # 모든 예측값과 레퍼런스 결합
    all_predictions_class = torch.cat(all_predictions_class, dim=0)
    all_references_class = torch.cat(all_references_class, dim=0)
    all_predictions_label = torch.cat(all_predictions_label, dim=0)
    all_references_label = torch.cat(all_references_label, dim=0)

    avg_val_loss = total_val_loss / len(dataloader_val)

    # 메트릭 계산
    metrics_class = compute_metrics(all_predictions_class, all_references_class, is_multilabel=False)
    metrics_label = compute_metrics(all_predictions_label, all_references_label, is_multilabel=True)

    # 각 메트릭을 'eval/' 및 'train/' 접두사와 함께 로깅
    metrics_class = {'eval_class/' + key: value for key, value in metrics_class.items()}
    metrics_label = {'eval_label/' + key: value for key, value in metrics_label.items()}
    metrics = {'train/loss': avg_train_loss, 'eval/loss': avg_val_loss}
    metrics.update(metrics_class)
    metrics.update(metrics_label)

    if epoch != 0 and epoch % 40 == 0:
        adjusted_epoch = epoch // 40
        metrics['train/epoch'] = adjusted_epoch
        wandb.log(metrics, step=adjusted_epoch)

    print({metric: f"{value:.4f}" for metric, value in metrics.items()})

# Example usage
# Train_and_Evaluate(dataloader_train, dataloader_val, device, model, loss_fn_multi_class, loss_fn_multi_label, optimizer)
```

# **6. Downstream Model Class**

```python
class DownstreamModel(nn.Module):
    def __init__(self, class_num, label_num, SIGMA):
        super(DownstreamModel, self).__init__()
        # SIGMA를 float으로 저장해두고 forward에서 GPU로 이동시킴
        self.SIGMA = SIGMA

        self.compress_layers = nn.ModuleList()
        for _ in range(5):
            layers = []
            layers.append(nn.Linear(2048, 1024))
            layers.append(nn.ReLU())
            layers.append(nn.Dropout(0.5))
            self.compress_layers.append(nn.Sequential(*layers))

        # 두 가지 분류 과제를 위한 두 개의 출력 레이어
        self.fc1 = nn.Linear(2097, 1024)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(1024, 256)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.5)

        # Multi-class classification output layer
        self.fc3_class = nn.Linear(256, class_num)

        # Multi-label classification output layer
        self.fc3_label = nn.Linear(256, label_num)

    def forward(self, input_l, input_b, input_r):
        batch_size = input_l.shape[0]

        # input_l 텐서를 첫 번째 차원을 기준으로 1 크기의 텐서로 분할
        split_tensors = torch.split(input_l, 1, dim=1)
        input = []

        # 분할된 텐서들을 순회
        for i, split_tensor in enumerate(split_tensors):
            # 각 split_tensor를 배치 크기에 맞게 2차원으로 재구성
            split_tensor = split_tensor.reshape(batch_size, -1)
            # 재구성된 텐서를 압축(compress) layer를 거쳐 변환
            input.append(self.compress_layers[i](split_tensor))

        # input_b(bert 임베딩)와 input_r(RoBERTa 임베딩)을 input에 추가
        input.append(input_b)
        input.append(input_r)
        input = torch.stack(input, dim=1)

        # X * X^T
        input_T = input.transpose(1, 2)
        input_P = torch.matmul(input, input_T)
        input_P = input_P.reshape(batch_size, -1)

        # PN func
        input_P = 2 * F.sigmoid(self.SIGMA * input_P) - 1

        a = torch.mean(input_l, dim=1)
        input = torch.cat([input_P, a], dim=1)

        # 공통 레이어 통과
        output = self.fc1(input)
        output = self.relu1(output)
        output = self.dropout1(output)
        output = self.fc2(output)
        output = self.relu2(output)
        output = self.dropout2(output)

        # Multi-class 및 Multi-label 분류 레이어를 통해 두 가지 예측 생성
        output_class = self.fc3_class(output)    # Multi-class output
        output_label = torch.sigmoid(self.fc3_label(output))  # Multi-label output (sigmoid로 각 레이블의 확률 출력)

        print(output_class)
        print(output_label)
        return output_class, output_label
```

# **8. 추론 메서드 정의**

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BertTokenizer, BertModel, RobertaTokenizer, RobertaModel
from transformers import AutoConfig

# Load the tokenizers and models for Llama2, BERT, and Roberta
llama2_tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf", use_auth_token="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", trust_remote_code=True)
llama2_tokenizer.pad_token = llama2_tokenizer.eos_token  # 패딩 토큰 설정
llama2_config = AutoConfig.from_pretrained("meta-llama/Llama-2-7b-chat-hf",use_auth_token="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", output_hidden_states=True)
llama2_model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",use_auth_token="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", config=llama2_config)

bert_tokenizer = BertTokenizer.from_pretrained('google-bert/bert-large-uncased')
bert_model = BertModel.from_pretrained('google-bert/bert-large-uncased')

roberta_tokenizer = RobertaTokenizer.from_pretrained('FacebookAI/roberta-large')
roberta_model = RobertaModel.from_pretrained('FacebookAI/roberta-large')

# Make sure all models are in evaluation mode and moved to the appropriate device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
llama2_model.eval().to(device)
bert_model.eval().to(device)
roberta_model.eval().to(device)

# Initialize the downstream model
class_num = 5  # For example, if you have 8 classes
SIGMA = 0.1  # SIGMA value for your downstream model
downstream_model = DownstreamModel(class_num, SIGMA).to(device)

model_load_path = "/content/drive/MyDrive/LLMEmbed/model_weights_stackexchange.pth"

# 가중치 로드
downstream_model.load_state_dict(torch.load(model_load_path, map_location=device))
downstream_model.eval()

def get_llama2_embedding(text, tokenizer, model, device):
    inputs = tokenizer(text, return_tensors='pt', max_length=128, padding="max_length", truncation=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        # Average the last 5 layers
        embedding = torch.stack([torch.mean(outputs.hidden_states[i], dim=1) for i in range(-1, -6, -1)], dim=1)
    return embedding

def get_bert_embedding(text, tokenizer, model, device):
    inputs = tokenizer(text, return_tensors='pt', max_length=128, padding="max_length", truncation=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        # Use pooler_output for BERT embeddings
        embedding = outputs.pooler_output
    return embedding

def get_roberta_embedding(text, tokenizer, model, device):
    inputs = tokenizer(text, return_tensors='pt', max_length=128, padding="max_length", truncation=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        # Use the first token ([CLS] token) representation
        embedding = outputs.last_hidden_state[:, 0, :]
    return embedding

def infer(text, downstream_model, device):
    # 각 모델로부터 임베딩을 추출
    llama2_emb = get_llama2_embedding(text, llama2_tokenizer, llama2_model, device)
    bert_emb = get_bert_embedding(text, bert_tokenizer, bert_model, device)
    roberta_emb = get_roberta_embedding(text, roberta_tokenizer, roberta_model, device)

    # Forward pass through the downstream model
    with torch.no_grad():
        prediction = downstream_model(llama2_emb, bert_emb, roberta_emb)

    return prediction
```
